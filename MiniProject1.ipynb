## Compute the number of schools and, using pandas, min, max, standard deviation and mean of the amount of grants that the Philadelphia School District received in 2010-11



A data frame is a container that can hold data of different types.  It is a two dimensional data structure with data organized in a tabular fashion.

A panda series is a list; a single element of data created from manipulating a data frame which allows you to use panda methods on those results.  The user can tell a series from a Data Frame because when the results are displayed the last line is not data - it is information about the data. It is also only a one dimensional array.

This exercise will use pandas to import data into a data frame.  Then the data will be modified so it can be manipulated.  The results for the Philadelphia school district will be isolated and summarized.

Step 1:  Import the school funding data as a dataframe using panda then look at the first 6 rows of the data using the head function.

```python
import pandas as pd
school_data = pd.read_csv('userssharedsdfschoolimprovement2010grants.csv')
school_data.head()
```

Step 2:  Look at the types of data in the data frame.

```python
school_data.dtypes
```

Step 3:  Rename the columns to eliminate spaces.

```python
school_data.columns =school_data.columns.str.replace(' ', '_')
school_data.columns
```

Step 4:  Change the name of the Amount field.

```python
school_data.rename(columns = {'2010/11/Award_Amount':'Amount'}, inplace=True)
school_data.columns
```

Step 5:  Remove the $ in the rows in the Amount field so that it can be converted to a number.

```python
school_data['Amount'] = school_data['Amount'].str.replace('$','')
school_data.head()
```

Step 6:  Change the Amount field to a number field.

```python
school_data['Amount'] = school_data.Amount.astype(float)
school_data.dtypes
```

Step 7: Filter on just the records for the Philadelphia City SD

```python
philly = school_data["District_Name"].str.contains("PHILADELPHIA CITY SD")
school_data[philly]
```

Step 8:  Calculate the number of schools in the Philadelphia City SD and min, max, mean and standard deviation of the funding.

```python
school_data[philly].describe()
```

There are 27 schools that received funding.  The minimum amount was 1.  The maximum amount was 980,283.  The mean amount was 555,936.52.  The standard deviation was 288,151.68.  Since I used only pandas functions, the standard deviation that pandas calculates is sample standard deviation.
